# Artist Based Lyric Categorization
A data mining approach to categorize song lyrics by their artist

## Overview  
This project attempts to create a model to classify song lyrics by artist. It uses a Multinomial Naive Bayes approach, where songs are treated as unordered collections of words. A variance threshold is applied in order to remove words that have low variance across artists from the corpus. The value of the variance threshold is determined by trying many values and calculating the accuracy using each one, and selecting the appropriate value. TFIDF (Term Frequency Inverse Document Frequency) is used to weight word counts such that words that are used less commonly used are weighed stronger. The average accuracy achieved with this method is approximately `0.65`~`0.70`.

## Data Retrieval

Here are the steps taken to retrieve the data from the web

##### myGenius<span></span>.py

##### parse_json.py
This file converts the `JSON` files output by `scrape.py` into a single `.txt` file that is read in by this notebook. It also does some preprocessing on the data, removing lyrics that are too short, as well as sections belonging to featured artists (which would pollute the data)


##### Note:
Creating `myGenius.py` to order results by popularity, as well as removing featured artist sections in `parse_json.py` gave an initial accuracy increase from approximately `0.49` to `0.63`



## Data Pre-processing
The data pre-processing first transforms the raw data into a form that can be processed into a model.
This involves 3 steps:  
1. Encode the labels into integer values
2. Separate the data into training and testing sets
3. Vectorize the lyrics so that each word in the lyric is mapped to a unique integer
4. Perform TF-IDF transform and Variance based feature selection

### Reading and Label Encoding
The cell below reads the data from the txt file in a csv form using a tab separator.
To perform this step the Python Panda Library is used.  

Here is a sample of what a section of "lyrics.txt" looks like ("..." indicating more lyrics that are edited out for sake of readability):  

```
Kendrick Lamar	   Compton, Compton, aint no city quite ... Be back, 15 minutes!  
Kendrick Lamar	   Ali, you thirsty on Twitter! You boo-boo, you ... we stand for it as if it’s as big as a religion  
Lauryn Hill	 Yo, remember back on the Bully When ... end that conversation with that, right?  
Lauryn Hill	 It could all be so simple ... Where were you when I needed you? Where were you?  
```
* The names on the left indicate the artist and the text on the right contains the lyrics for a song written by the artists.  
* The two sections are separated by a tab ("\t") character which acts as the separator for the Panda read_csv function.  
* The file is separated by tabs instead of commas because the song lyrics occasionally contain commas, but never contain tabs.  
* This separation simplified the parsing without sacrificing readability of the data.  

In this cell the class labels are also converted into integers using the `LabelEncoder` from sklearn

````python
import pandas as pd
from sklearn import preprocessing

data = pd.read_csv('./lyrics.txt',sep='\t',names=['artist','title','lyrics'])

labeler = preprocessing.LabelEncoder()
labeler.fit(data['artist'])

artist_array = labeler.classes_
label_map = dict(zip(labeler.classes_,labeler.transform(labeler.classes_)))

encoded_artists = labeler.transform(data['artist'])

print(f"Number or artists/classes: {len(artist_array)}")
data.head(-1)
````

### Creating Testing and Training Sets
In this section the data is separated into testing and training sets. Where the testing set is 40% of the total dataset.  

To perform the split the `train_test_split` function from sklearn is used. Using the `encoded_artists` parameter the function was set to randomly select 40% of the data to be used as the test set.  
Each time a new pair of sets is generated the distribution of instances is different.  
This element of randomness helps to prevent overfitting of the model.  

When comparing models the randomness was removed by adding the parameter `random_state = 1`.  
The `random_state` parameter is used as the seed for the random number generator. Thus, making the split deterministic,

````python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(data['lyrics'],encoded_artists,test_size=0.4)
````

### Encoding of the lyric data
This code transforms the lyrics into an integer vector.

To perform the transformation the `CountVectorizer` class from sklearn is used.  
This object reads in the dataset and encodes the terms of each document into a vector of integers.  
In other words, given a data set with vocabulary *V*, the `CountVectorizer` creates |*V*| features where each feature is a unique term in the corpus.  
The `CountVecotizer` then fills each row in the matrix with the count of each term in the corresponding document.

````python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
vec = CountVectorizer()

x_train_v = vec.fit_transform(x_train)
x_test_v = vec.transform(x_test)

y_train_v= np.int32(np.array(y_train))
y_test_v = np.int32(np.array(y_test))
````

## Variance Thresholding, TF-IDF, Multinomial Bayes
In the following sections the dataset is run through a Variance Threshold feature selection and a TF-IDF transformer.  
Then the prepared data is fed into the Multinomial Naive Bayes Model.  

### Variance Thresholding [1]

The data is first reduced by using a Variance Threshold feature selection using the `VarianceThreshold` class from sklearn.  
This method of feature selection calculates the variance of each feature in the dataset.  
Using the adjustable threshold value, it removes features that have a variance below the threshold.  
In other words, it removes common terms in the dataset that are shared by many documents.  

In the following example data set, all the instances share a value of 0 for the first feature.
This means that feature 1  has a variance of 0. The `VarianceThreshold` then elimiates the feature from the dataset.


````python
from sklearn.feature_selection import VarianceThreshold
# This is an example to illustrate how Variance Threshold works
EXAMPLE_DATA = [[0, 2, 0, 3], [0, 1, 4, 3], [0, 1, 1, 3]]
EXAMPLE_SELECTOR = VarianceThreshold(threshold=0)
EXAMPLE_SELECTOR.fit_transform(EXAMPLE_DATA)
````

### TF-IDF [2]
After the Variance Threshold feature selection, the data is weighted by using the TF-IDF transformer using the `TfidfTransformer` class from sklearn.  

TF-IDF scales the weights of the term counts based on inverse frequency of the term in the entire document.

### Finding the Variance Threshold

In the next cell, the model is trained and evaluated on the training set with various variance threshold values.  
The graph at the bottom of the cell displays the improvement of the result as the threshold becomes smaller.  

The optimized threshold value is saved to be used in the final model.


````python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

import matplotlib.pyplot as plt

from sklearn.feature_extraction.text import TfidfTransformer


mnb = MultinomialNB(alpha=1.0)
tf = TfidfTransformer(norm=None)

def pred(thresh):
    vt = VarianceThreshold(threshold=thresh)
    x_train_tf = vt.fit_transform(x_train_v)
    x_test_tf = vt.transform(x_test_v)
    x_train_vt = tf.fit_transform(x_train_tf)
    x_test_vt = tf.transform(x_test_tf)
    mnb.fit(x_train_vt,y_train_v)
    y_pred = mnb.predict(x_train_vt)
    return accuracy_score(y_train_v,y_pred)

it = 500
best_p = 0
best_th = None
p = []
for i in range(it):
    curr = pred(1/(i+1))
    p.append(curr)
    if(curr>best_p):
        best_p=curr
        best_th = 1/(i+1)
plt.rcParams["figure.figsize"] = [6,6]
plt.xlabel('inverse of variance threshold')
plt.ylabel('accuracy')
plt.plot(p)
plt.show()
print(f"Best Variance Threshold Value for training set: {best_th}")
````

## Results
The results of our model proved to have an accuracy of roughly 65-70%.  
Considering there are 32 classes and the accuracy of a random classifier would be roughly 3%, it is a remarkably effective.


Below the results of both using Variance Threshold and skipping Variance Threshold are displayed.  

The `accuracy_score` displays the ratio of correct predictions divided by the total number of test instances.

````python
#Without Variance Thresholding
x_train_tf = tf.fit_transform(x_train_v)
x_test_tf = tf.transform(x_test_v)

mnb.fit(x_train_tf,y_train_v)
y_pred = mnb.predict(x_test_tf)
print(f"Accuracy without Variance Threholding: {accuracy_score(y_test,y_pred)}")

#With Variance Thresholding
vt = VarianceThreshold(threshold=0.004444)
x_train_vt = vt.fit_transform(x_train_v)
x_test_vt = vt.transform(x_test_v)

x_train_tf = tf.fit_transform(x_train_vt)
x_test_tf = tf.transform(x_test_vt)

mnb.fit(x_train_tf,y_train_v)
y_pred = mnb.predict(x_test_tf)
print(f"Accuracy with Variance Threholding: {accuracy_score(y_test,y_pred)}")

````

## Analysis
Below are some analytical insights that were drawn from the model.  

This section includes 3 sections: A look at the top 5 predictions, the most used words for each artist, and a confusion matrix.

### Runners Up

Here misclassified songs are investigated.  
The table below displays the top 5 predicted artists to see if the correct artist appears. This gives a measure of how close the algorithm is to finding the correct result. If songs do not have the correct label in the top 5 predictions, it is also possible that the song itself is an outlier.  

The statitics shows that the majority of misclassified songs are nearly correct when considering the top 5 predictions.  

````python
predicted_artists = []

artist_n = len(artist_array)
artist_n_list = [x for x in range(artist_n)]

top_data = {'true':[]}

limit = 5

missclassified_n = 0
correct_in_limit = 0

for i in range(limit):
    top_data[i+1] = []
for test_i in range(x_test_vt.shape[0]):
    predicted_5_artists  =" "
    true_value = y_test[test_i]

    # this is an array of probabilities for 'test_i'th song
    probabilities_by_classes = mnb.predict_log_proba(x_test_vt[test_i])[0]

    class_zip = list(zip(artist_n_list,probabilities_by_classes))

    # sort the predictions in descending order so that the class in the 1st position is the prediction
    class_zip.sort(key=lambda x:x[1],reverse=True)

    #check if top prediction == true value i.e the model predicted the correct class
    if(class_zip[0][0]==true_value):
        continue

    missclassified_n += 1

    t = str(artist_array[true_value])
    top_data['true'].append(t)

    for i,artist in enumerate(class_zip):
        #increments if correct value is in limit
        if(artist[0]==true_value):
            correct_in_limit +=1
        if(i>=limit):
            break
        top_data[i+1].append(artist_array[artist[0]])
        predicted_5_artists = predicted_5_artists + "," +artist_array[artist[0]]
    predicted_artists.append(predicted_5_artists)

print(f"missclassified instances: {missclassified_n}\n\
missclassified but correct prediction in top {limit}: {correct_in_limit}\n\
% of missclassified but in top {limit}: {correct_in_limit/missclassified_n}")
    
pd.DataFrame(top_data)
````

### Top Words
Here we investigate the most common words for each artist.  
In this analysis, a stop word filter is used because it focusses the analysis on the unique words that distinguish an artist.  
This analysis gives us insight into what kind of words we can use to try and identify a particular artist, and gives another measure of spread and variance.


````python
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english")).union({'like', 'im','got'})

weights = mnb.coef_
weights = vt.inverse_transform(weights)
samples_n = mnb.class_count_

tw_data = {'artist':[],'song count':[]}

limit = 30
for i in range(limit):
    tw_data[i+1] = []

for artist_i in range(len(weights)):
    tw_data['artist'].append(artist_array[artist_i])
    tw_data['song count'].append(samples_n[artist_i])
    
    #combines words (vec.get_feature_names) with weight of each word (ar[artist_i])
    words_weight = list(zip(vec.get_feature_names(),weights[artist_i]))

    #Sorts the words by their weights
    words_weight.sort(key = lambda x:x[1],reverse=True)

    count=0
    top_words = []
    for _,word in enumerate(words_weight):

        #if word[1]==0 then the word was not used by the artist
        if(word[1]!=0 and word[0] not in stop_words):
            count+=1
            top_words.append(word[0])
            tw_data[count].append(word[0])
            if(count>=30):
                break
pd.options.display.max_columns = 40
pd.DataFrame(tw_data)
````

### Confusion Matrix
Here the confusion matrix is constructed. This is one of the most interesting parts of the analysis. Here you can see the results, but also some information about the weaknesses of the algorithm, and the structure of the dataset. The confusion matrix shows the correct classes on the vertical, and the predicted classes on the horizontal. The numbers shown are proportional to the number of songs belonging to the artist on the vertical that were predicted to belong to the artist on the horizontal. Therefore, correct predictions appear on the diagonal. As seen below, there is a strong diagonal trend, however there is a fair amount of unwanted noise on the matrix. 

There are some horizontal and vertical trends in mispredictions, which indicate that certain artists are often misclassified or misattributed to other songs, respectively. Further analysis could be done to investigate what properties make these artists prone to creating bad predictions.


````python
import itertools

from sklearn.metrics import confusion_matrix


# This plotting method was taken from [4]:
# https://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html#sphx-glr-auto-examples-model-selection-plot-confusion-matrix-py


def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=90)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')


# Compute confusion matrix
cnf_matrix = confusion_matrix(y_test, y_pred)
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plt.rcParams["figure.figsize"] = [25,25]
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=[x for x,_ in label_map.items()],
                      title='Confusion matrix, without normalization')

plt.show()
````

## Discussion

This section is a discussion which justifies the decisions that were made in designing the model and outlines some of the challenges that were encountered.
The sections are explored in chronological order of development.

### Variance Threshold vs Stop Words

Basic feature selection consists of removing irrelevant or noise features. In this case, removing words that do not meaningfully identify a particualar artist.  

The initial approach was using a list of stop words as a filter. This method uses a list of common words that are essentially "filler" (e.g. 'the','a','is'...). Then it filters the dataset and removes the common words. 

This method works, but many stop word lists are based on regular english text bodies. This model required a special filter due to the eccentricity of the text.  
Compared to a conventianal text, this dataset had common words that are essentially stop words that do not appear on the list, or words in the stop words list that are very infrequently used. In order to apply a more dynamic word filter, `VarianceThreshold` was used.

The variance threshold works by calculating the variance across all documents for each word, and removing all occurences of words with a variance below a threshold value. The next step is to figure out how to determine the threshold value. This was done by sampling at many different threshold values, and calculating the accuracy for each one, and choosing the best value for the current dataset. This can be graphed, as seen above. The initial variance graphs had a clear peak area, which indicates that there is an ideal variance threshold specific to each training dataset.

Using a stop word filter resulted in accuracies around `0.20` - `0.25`. For these test a stop word lists from `nltk` and `sklearn` were tested with similar results.  
Adding the variance threshold initially increased the accuracy increased from around `0.30`-`0.35` to `0.45`-`0.49`.

After adding TF-IDF after the variance threshold (explained below), the shape of the graph changed significantly. Very low thresholds work better, although still providing better accuracy than no threshold.

### Data Sanitization

Initially, the basic data sanitization setup was to scrape lyrics data from `Genius.com` using the lyricsgenius package, which is a python wrapper for the API. lyricsgenius also uses the Genius API to search for songs and artists, but the API does not provide access to actual song lyrics, so it also uses the BeautifulSoup library to scrape the lyrics and deliver json files. After scraping the lyrics, our `parse_json.py` script was used to turn multiple json files into a single text file of tab-separated values, removing non-content data (such as labels) from songs, and also remove any very short or invalid songs. 

Using this process, the accuracies were approximately `0.30`-`0.35` without variance thresholding, and approximately `0.45`-`0.49` with variance threshold.

In order to increase the accuracy a few techniques were applied. Firstly, Many songs contain featured verses, which are lyrics belonging to artists other than the primary artist, this will clearly impact the accuracy poorly. In order to remedy this, the `parse_json.py` script was modified using a regular expression to isolate song sections (verse, bridge, chorus, hook...), and check the section labels before removing them to see whether that section is performed by the target artist, removing sections that are not.

Secondly, the `Genius.search_artist()` function in the lyricsgenius package was used to retrieve a fixed number songs for a particular artist, however this function only returns songs in alphabetical order. So for example, a request for 50 songs from a particular artist will return the 50 first songs by alphabetical order. This means that the quality of content in these songs is essentially random. Many may not be actual songs, but interludes, spoken pieces, or notes of some form, and some are foreign translations of songs into other languages. Even if a result is a "real" song, it may be a very obscure song that is not reflective of the artist's normal style.

Although the `Genius.search_artist()` function does not allow searches to be ordered by anything other than alphabetical, the underlying Genius API does support ordering by popularity as well. In order to access this functionality, a file called `myGenius.py` was created, containing the `MyGenius` class. This class is a subclass of the `Genius` class in the lyricsgenius package, and overrides the `search_artist()` function with a version that allows for search results to be sorted by popularity. Popularity is a better sort criteria than alphabetical for this application, because the most popular songs are more likely to have a consistent vocabulary and style, and do not include the low-quality content mentioned above.

This updated process that removes featured artist content and sorts by popularity was able to increase popularity from around `0.49` to around `0.63` at best.

### TF-IDF

This project analyzes hip-hop lyrics, under the assumption that artists in this genre tend to use very unique and identifying words often, this can be seen in the earlier table of frequent words. Artists will often use their own name or nickname, the name of their city or neighbourhood, record label, or just unique sounds and words. In the analysis it is seen that there are a large number of words that are uniquely used by a very small subset of artists. In order to leverage this feature of the dataset, TF-IDF is used.

TF-IDF works by weighting the word counts such that words that have a high ratio of appearances in a song compared to overall appearances are weighted stronger, and vice versa. This allows these unique words to have a stronger effect on the classification algorithm. It was found to be important not to allow normalization in the TF-IDF algorithm, as that would remove the severity of differently weighted word counts.

Initially the TF-IDF transformer had a negative effect on the result. It was discovered that by turning off the default normalization used in TF-IDF in `sklearn` the results showed improvement. This is likely due to the sparseness of feature values and the unbalanced distribution of features. The true reasoning for this will have to be researched more in the future.  

Using TF-IDF increased the accuracy from around `0.63` to up to around `0.65-0.70`


## Conclusions

Overall the model proved to have an accuracy of roughly 65-70% on the acquired dataset using a 60-40 training testing split.  
The project also provided insight into the similarities between artists and the variety of words that can be found in the domain of artists that we chose.

### Future Work

This section discusses the aspects of the project that need further research and development.  
For the evaluation, a more robust evaluation method could be implmented such as k-fold validation. The current accuracy metric is somewhat insightful, but it is not as rigorous a k-fold validation which basically performs the current testing method k times.  

Using the analysis that has been done, better preprocessing algorithms can be investigated.  By looking at the top 5 predictions it is possible to identify outlying song lyrics and analyze the factors that causes the outlier and possibly further sanitize the data or amend the model. A similar analysis can be done with the confusion matrix. By identifying commonly miscatagorized artists, further research can be done to improve the model for the incorrectly identying classes.  

A main concern is that the TF-IDF and Variance Threshold do not take into consideration the training labels, and cannot weight words counts based on their uniqueness to a particular class. Perhaps further weighting the training data based on the training labels before fitting with Multinomial Naive Bayes with the aim of leveraging those highly unique words would lead to higher accuracy.
